# using sqlite search

```python
from sqlitesearch import TextSearchIndex

sqlite_index = TextSearchIndex(
    text_fields=["question", "section", "answer"],
    keyword_fields=["course"],
    db_path="faq.db"
)
```

```python
sqlite_index.search("i have just discovered the course, can i join it?")
```

```python
assistant = RAGBase(sqlite_index, openai_client)
```

```python
assistant.rag("i have just discovered the course, can i join it?")
```
